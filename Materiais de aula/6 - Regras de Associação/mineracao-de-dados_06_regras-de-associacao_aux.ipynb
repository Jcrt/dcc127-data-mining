```
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder

transacoes = [["biscoito","cerveja","chá","salaminho"],
["cerveja","couve","linguiça","pão","queijo"],
["café","brócolis","couve","pão"],
["brócolis","café","cerveja","couve","pão","salaminho"],
["brócolis","café","couve","pão","refrigerante"],
["couve","linguiça"]]

te = TransactionEncoder()
dadosTransformados = pd.DataFrame(te.fit(transacoes).transform(transacoes), columns=te.columns_)

print(dadosTransformados)

itemsets = apriori(dadosTransformados, min_support=0.3, use_colnames=True)

itemsets
```

```
from mlxtend.frequent_patterns import association_rules

regras = association_rules(itemsets, metric="confidence", min_threshold=0.7)

regras.sort_values(by="lift",  ascending=False)

```
