# Exemplos de padronização/normalização dos dados

```
from sklearn import preprocessing

sub = dados[ [ "Irmãos", "Salário" ] ].dropna()
print(sub)
```

```
padronizados = pd.DataFrame( preprocessing.scale( sub ), columns=sub.columns )
print(padronizados)

print("Média")
print(padronizados.mean())
print("Desvio padrão")
print(padronizados.std())
```

```
scaler = preprocessing.StandardScaler()
padronizados = pd.DataFrame( scaler.fit_transform( sub ) )
print(padronizados)
```

```
scaler = preprocessing.MinMaxScaler()
normalizados = pd.DataFrame( scaler.fit_transform(sub), columns=sub.columns )
print(normalizados)
print(normalizados.min())
print(normalizados.max())
```

```
scaler = preprocessing.MinMaxScaler( feature_range=(-1, 1) )
normalizados = pd.DataFrame( scaler.fit_transform(sub), columns=sub.columns )
print(normalizados)
print(normalizados.min())
print(normalizados.max())
```

```
scaler = preprocessing.RobustScaler()
normalizados = pd.DataFrame( scaler.fit_transform(sub), columns=sub.columns )
print(normalizados)
```

# Tratamento de variáveis categóricas/nominais

```
from sklearn.preprocessing import LabelBinarizer

sub = dados[["Passatempo"]].replace({float('nan'):'passatempo-nan'})
print(sub)

colunas = sub["Passatempo"].unique()
colunas.sort()
print(colunas)

lb = LabelBinarizer()
lb.fit(colunas)

novosDados = pd.DataFrame( lb.transform( sub ), columns=colunas )
print(novosDados)
print(novosDados[["Dança", "Cinema", "Teatro"]])
```

```
sub = dados[["Passatempo"]]
novosDados = pd.get_dummies( sub )
print(novosDados)
```

# Tratamento de dados faltantes

```
sub = dados[["Idade", "Salário"]]
print(sub)

print(sub.isnull())
print(sub.isnull().sum())
```

```
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=float('nan'), strategy='mean')
imp.fit( sub )

dadosCompletos = pd.DataFrame( imp.transform( sub ), columns=sub.columns )
print(dadosCompletos)

d = pd.concat( [sub, dadosCompletos], axis=1 )
print(d)
```

```
sub = dados[["Escolaridade", "Salário"]]
print(sub)

sub = sub.dropna(subset=["Escolaridade"])
print(sub)
print(sub.isnull().sum())
```

```
salarioMediana = sub.groupby("Escolaridade").median()
print(salarioMediana)

for e in salarioMediana.index:
  sub.loc[ sub["Escolaridade"]==e ] = sub.loc[ sub["Escolaridade"]==e ].fillna( salarioMediana["Salário"].get(e) )

print(sub)
```

# Detecção de outliers:

http://scikit-learn.org/stable/modules/outlier_detection.html

```
sub = dados[["Irmãos", "Salário"]].dropna()
print(sub)

sub["Irmãos"][9] = 7
sub["Salário"][9] = 90000
print(sub)
```

```
import numpy as np

print( sub[ np.abs(sub-sub.mean()) <= 3*sub.std() ] )
```

```
print( sub[ np.abs(sub-sub.mean()) <= 2*sub.std() ] )
```

```
print( sub[ np.abs(sub-sub.mean()) <= sub.std() ] )
```

```
print( sub["Salário"][ np.abs(sub["Salário"]-sub["Salário"].mean()) <= 2*sub["Salário"].std() ] )
```

```
q1, q3 = np.percentile(sub["Salário"], [25, 75])
print(q1)
print(q3)
```

```
iiq = q3-q1
print(iiq)
```

```
min = q1-iiq*1.5
max = q3+iiq*1.5
print(min)
print(max)
```

```
print( sub["Salário"][ (dados["Salário"] < min) | (dados["Salário"] > max) ] )
```
