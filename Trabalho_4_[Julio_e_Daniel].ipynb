```
from google.colab import drive
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```
drive.mount('/content/drive')
dados = pd.read_csv("/content/drive/MyDrive/trabalho4_dados_07.csv")

dados
```

# Suporte mínimo
****
Para calcular o suporte mínimo, iniciamos os testes com 0.1, e isso significa que os produtos devem estar presentes em 10% da base de dados, o que nos retornou quatro produtos isolados. 

```
itemsets = apriori(dados, min_support=0.1, use_colnames=True)

itemsets
```

# Ajuste de suporte mínimo
****
Fomos reduzindo o suporte mínimo até chegarmos em um número que julgamos interessante. Com o suporte mínimo em 0.007, a regra de suporte mínimo exigiria que os ítens estivessem na mesma compra por, pelo menos, 40 vezes para serem uma possível combinação frequente.

```
itemsets = apriori(dados, min_support=0.017, use_colnames=True)

itemsets
```

# Verificação de suporte
****
Efetuamos alguns testes com as combinações geradas no *step* anterior e, para as instâncias geradas, todas geraram um número igual ou maior a 95

```
d1 = dados[(dados["CREME"] == 1) & (dados["PROD_LIMPEZA"] == 1) & (dados["CAFE"] == 1)]
print("Total de linhas:", len(d1.index))
```

#Descobrindo regras de associação
****
Para descobrir as regras de associação, iniciamos o limiar de confiança mínima em 0.5 e foram retornadas muitas regras de associação. 

```
regras = association_rules(itemsets, metric="confidence", min_threshold=0.5)

print("Total de regras de associação:", len(regras.index))
regras.sort_values(by="lift",  ascending=False)
```

#Ajustando as regras de associação
****
Como foram geradas muitas regras de associação, aumentamos o limiar mínimo para 0.99, o que nos gerou cerca de 270 ítens com regras de associação. 
Isso nos revela que todas as regras de associação descobertas são fortes.

```
regras = association_rules(itemsets, metric="confidence", min_threshold=0.99)

print("Total de regras de associação:", len(regras.index))
regras.sort_values(by="lift",  ascending=False)
```
