```
def testeDeFuncao(x):
  return x**2, math.sqrt(x)

a, b = testeDeFuncao( 4 )
print(a)
print(b)

c = testeDeFuncao( 9 )
print(c)
print(c[0])

print(type(c))
c = 1
print(type(c))
```
